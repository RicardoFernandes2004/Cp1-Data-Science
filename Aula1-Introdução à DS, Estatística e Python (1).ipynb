# Introdução à Ciência de Dados
A ciência de dados é um campo interdisciplinar que utiliza métodos, processos, algoritmos e sistemas científicos para extrair conhecimento e insights de dados estruturados e não estruturados. Ela combina várias disciplinas, incluindo estatística, matemática, ciência da computação e engenharia da informação.

**Definições iniciais**

```
Dados: Conjunto de valores quantitativos ou qualitativos
```

```
Informação: Dados processados e organizados de maneira que tenham significado
```

```
Conhecimento: Informação aplicada e utilizada para tomar decisões
```

**Objetivos**


* Extrair informações valiosas dos dados.
* Ajudar na tomada de decisões baseada em dados.
* Melhorar processos e operações com base em *insights* de dados.

**Laços de repetição**

**Listas**

````python
# Criação e manipulação de listas
lista = [1, 2, 3, 4, 5]

# Acessando elementos
print(lista[0])  # Primeiro elemento
print(lista[-1])  # Último elemento

# Adicionando e removendo elementos
lista.append(6)  # Adiciona 6 ao final da lista
print(lista)
lista.remove(3)  # Remove o primeiro 3 encontrado
print(lista)
````

**Dicionários**

````python
# Criação e manipulação de dicionários
dicionario = {"nome": ["João","Michelle"], "idade": [25,27], "cidade": ["São Paulo","Porto Alegre"]}

# Acessando valores
print(dicionario["nome"])
print(dicionario.get("idade"))

print(dicionario)
# Adicionando e removendo pares chave-valor
dicionario["email"] = ["joao@example.com",'mi@chelle.com']
print(dicionario)
del dicionario["cidade"]

print(dicionario)
````

**Funções**

````python
# Definindo uma função
def saudacao(nome):
    return f"Olá, {nome}!"

# Chamando a função
mensagem = saudacao("Maria")
print(mensagem)
````

**Manipulação de strings**

````python
texto = "Ciência de Dados"

# Convertendo para maiúsculas e minúsculas
print(texto.upper())
print(texto.lower())

# Fatiando strings
print(texto[:7])  # "Ciência"
print(texto[8:])  # "de Dados"

# Substituindo partes da string
novo_texto = texto.replace("Dados", "Informação")
print(novo_texto)
````

## scikit-learn

scikit-learn é uma das bibliotecas mais populares para machine learning em Python. Ela fornece ferramentas simples e eficientes para análise de dados e modelagem preditiva.

Principais Funcionalidades:

    Classificação: K-Nearest Neighbors, SVM, Decision Trees, Random Forest.
    Regressão: Linear Regression, Ridge, Lasso, ElasticNet.
    Clustering: K-Means, DBSCAN, Hierarchical clustering.
    Redução de Dimensionalidade: PCA, ICA, LDA.
    Pré-processamento de Dados: StandardScaler, MinMaxScaler, Normalizer.

# TensorFlow

TensorFlow é uma biblioteca open-source desenvolvida pelo Google para computação numérica e machine learning. É amplamente usada para construir e treinar redes neurais profundas.

Principais Funcionalidades:

    Redes Neurais: Construção e treinamento de redes neurais profundas e convolucionais.
    Machine Learning: Suporte a modelos clássicos de machine learning.
    Desenvolvimento Mobile: Suporte para TensorFlow Lite para dispositivos móveis.
    Distribuição e Desempenho: Suporte para treinamento distribuído e aceleração de hardware com GPUs e TPUs.

# Keras

Keras é uma API de alto nível para construir e treinar modelos de deep learning. Pode ser utilizada em conjunto com TensorFlow, Theano ou CNTK como backend.

Principais Funcionalidades:

    Modelos Sequenciais: Criação fácil de modelos de deep learning sequenciais.
    Modelos Funcionais: Criação de modelos complexos com múltiplas entradas e saídas.
    Camadas Pré-definidas: Uma vasta coleção de camadas, incluindo convolucionais, recursivas, de pooling, etc.

# PyTorch

PyTorch é uma biblioteca de machine learning desenvolvida pelo Facebook. É popular para pesquisa e desenvolvimento em deep learning devido à sua flexibilidade e facilidade de uso.

Principais Funcionalidades:

    Autograd: Diferenciação automática para construção e treinamento de redes neurais.
    TorchScript: Transforma modelos PyTorch em modelos otimizados e independentes.
    Distribuição: Suporte para treinamento distribuído e aceleração de hardware.
